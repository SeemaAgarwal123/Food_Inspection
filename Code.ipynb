### The categories which represent a restaurant are Restaurant, Snack Bar and Mobile Unit because they represent a restaurant most closely.

```python
def result(x):                                                               # function to put 1 where the category is among the three defined categories
    if x['Category']=='Restaurant':
        return 1
    elif x['Category']=='Snack Bar':
        return 1
    elif x['Category']=='Mobile Unit':
        return 1
    else:
        return 0
df['Is_restaurant']=df.apply(lambda x : result(x),axis=1)             # adding the column 'Is_restaurant' to the dataframe

df
```

### Step 1.2

```python
import datetime as dt
df['Inspection Date'] =  pd.to_datetime(df['Inspection Date'])           # Converting the Inspection_date column into a datetime column
df['Month'] = df['Inspection Date'].apply(lambda x: x.strftime('%B'))    # new column for month of inspection
df['Year'] = df['Inspection Date'].apply(lambda x: x.strftime('%Y'))     # new column for year of inspection
df['Month_Year'] = pd.to_datetime(df['Inspection Date']).dt.to_period('M') # new column for month and year of inspection
df
```
